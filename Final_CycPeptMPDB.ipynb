```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### The data we are importing already has some deleted columns

```python
df = pd.read_csv("CycPeptMPDB_peptide.csv")
df
```

### Printing all column names

```python
for column in df:
    print (column)
```

### Data cleaning

```python
df.drop(['PAMPA', 'Caco2', 'MDCK','RRCK'], axis = 1, inplace = True)
```

```python
df = df.dropna()
```

```python
df.drop(['SMILES', 'Structurally_Unique_ID', 'Original_Name_in_Source_Literature', 'Year', 'Source', 'CycPeptMPDB_ID'], axis = 1, inplace = True)
```

```python
sns.displot(df['Permeability'], color = 'green', height=5, aspect=2)
```

### Removing the outliers

```python
df = df[df.Permeability != -10]
```

```python
sns.displot(df['Permeability'], color = 'green', height=5, aspect=2)
```

```python
for column in df:
    print(df.info(['column']))
```

```python
df1 = pd.get_dummies(df['Molecule_Shape'])
df = pd.concat([df, df1], axis=1).reindex(df.index)
df.drop('Molecule_Shape', axis=1, inplace=True)
```

```python
df.head(30)
```

```python
df.drop('Lariat', axis = 1, inplace = True)
```

```python
# correlations_dict = {}
for column in df:
    print(column,  df['Permeability'].corr(df[column]))
```

```python
sns.scatterplot(df, y = 'Permeability', x = 'NumAliphaticCarbocycles')
```

```python
sns.scatterplot(df, y = 'Permeability', x = 'NumSaturatedCarbocycles')
```

### Convert permeability Column into Boolean [1 if value is good; 0 otherwise]

```python
df['Permeability'] = df['Permeability'].apply(lambda x: 1 if x >= -6 else 0)
```

```python
# df.head(30)
```

```python
sns.countplot(df, x = 'Permeability')
```

```python
x = df.drop(['Permeability'], axis = 1)
```

```python
y = df['Permeability']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42)

scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
ytest.shape, xtest.shape, xtrain.shape, ytrain.shape
```

### Recursive feature elimination

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
```

```python
rfe = RFE(estimator=DecisionTreeClassifier(), n_features_to_select=10)
rfe.fit(x, y)
```

```python
for i, col in zip(range(x.shape[1]), x.columns):
    print(f"{col} selected={rfe.support_[i]} rank={rfe.ranking_[i]}")
```

```python
rfe = RFE(estimator=XGBClassifier(), n_features_to_select=10)
rfe.fit(x, y)
```

```python
for i, col in zip(range(x.shape[1]), x.columns):
    print(f"{col} selected={rfe.support_[i]} rank={rfe.ranking_[i]}")
```

### XGB MODEL WITH ALL FEATURES

```python
y0 = df['Permeability']
```

```python
x0 = df.drop(['Permeability'], axis = 1)
```

```python
from xgboost import XGBClassifier
```

```python
params = {
    "learning_rate": 0.5
}

model1 = XGBClassifier(**params)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x0, y0, test_size = 0.2, random_state = 42)

scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
model1.fit(xtrain, ytrain)
```

```python
ypred = model1.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score
```

```python
accuracy = accuracy_score(ytest, ypred)
accuracy
```

```python
print(classification_report(ytest, ypred))
```
